```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Загрузка данных
data = pd.read_csv("data/price_data.csv")


plt.figure(figsize=(15, 6))

# Построение графика
data['timestamp'] = pd.to_datetime(data['timestamp'])
data.set_index('timestamp', inplace=True)
data.groupby('source')['price'].plot(legend=True, title="Динамика цен AAVE/USDT на CEX и DEX")

plt.show()
```

```python
data[data['source'] == 'DEX (Uniswap)'].shape[0], data[data['source'] == 'CEX (Binance)'].shape[0]
```

```python
# Создание графиков
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Построение графиков для CEX и DEX
data[data['source'] == 'DEX (Uniswap)']['price'].plot(ax=ax[0])
data[data['source'] == 'CEX (Binance)']['price'].plot(ax=ax[1])

# Установка заголовков
ax[0].set_title("DEX (Uniswap) - цены AAVE/USDT")
ax[1].set_title("CEX (Binance) - цены AAVE/USDT")

# Получение общих пределов для оси y
y_min = min(data['price'])
y_max = max(data['price'])

# Применение одинаковых пределов y для обоих графиков
ax[0].set_ylim(y_min * 0.9999, y_max * 1.0001)
ax[1].set_ylim(y_min * 0.9999, y_max * 1.0001)

plt.tight_layout()
plt.show()
```
